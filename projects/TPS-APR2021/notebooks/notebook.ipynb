```python
sub = pd.read_csv("D:/Documents/GitHub/ml-pipeline/project/TPS-APR2021/submission_LOGIT_CL.csv")
sub.head()
```

```python
test_preds = sub.Survived.values.tolist()
```

```python
threshold = 0.5
```

```python
final_int_preds = []
for proba in test_preds:
    if proba >= threshold:
        pred_int = 1
    else:
        pred_int = 0
    final_int_preds.append(pred_int)
```
